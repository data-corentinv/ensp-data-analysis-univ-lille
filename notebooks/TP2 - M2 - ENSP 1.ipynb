```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,10)
y_squared = x**2
y1 = 2* x**2 + 3 
y2 = -x**2
```

```python
fig, ax = plt.subplots(1,1, figsize=(4,3))
ax.plot(x, y1, label="y1")
ax.plot(x, y_squared, label="y_squared")
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title('Title')
ax.legend()
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,3))
ax.plot(x,y_squared, 'xb', label='x**2')
ax.plot(x,y1, 'or', label='2x**2+3')
ax.plot(x,y2, '+g', label='-x**2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Mon premier graphique')
ax.legend()
ax.grid(True)
```
