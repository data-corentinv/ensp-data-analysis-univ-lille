```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid-hospit-2023-01-23-19h03.csv', sep = ";")
df_clean = df[df.sexe==0]
```

```python
mask = df_clean.jour.isin(['2023-01-23','2023-01-22','2023-01-21'])
resultat = df_clean[mask]\
             .groupby(["dep"])\
             .agg({"hosp": sum})\
             .reset_index()
```

```python
# ex1 
resultat.plot("dep","hosp", kind='bar', figsize=(15,3), grid=True)
```

```python
# ex2
x = resultat["dep"]
y = resultat["hosp"]

fig, ax = plt.subplots(1,1, figsize=(15,5))
ax.bar(x, y)
ax.set_xlabel("dep")
ax.set_ylabel("nb hosp")
ax.set_title("Nombre dhosp sur les 3 derniers jours en fonction des départements")
ax.set_xticklabels(x, rotation=90)
ax.grid(True)
```
